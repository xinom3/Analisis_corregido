```python
import pandas as pd
import numpy as np

# Cargar los datos (ACTUALIZADOS)
jose_luis = pd.read_csv('/home/xinome/Documentos/Analisis_corregido/datos/libreta_jose_luis.csv')
maria_elena = pd.read_csv('/home/xinome/Documentos/Analisis_corregido/datos/libreta_maria_elena.csv')
mari_jose_luis = pd.read_csv('/home/xinome/Documentos/Analisis_corregido/datos/mari_libreta_jose_luis.csv')
mari_maria_elena = pd.read_csv('/home/xinome/Documentos/Analisis_corregido/datos/mari_libreta_maria_elena.csv')

print("=== JOSÃ‰ LUIS (ACTUALIZADO) ===")
print("Registros de JosÃ© Luis:")
print(jose_luis[['Monto', 'Fecha de prestamo', 'Abonos', 'Notas']].to_string())
```

```python
# Limpiar montos y convertir a nÃºmeros
def limpiar_monto(monto):
    if isinstance(monto, str):
        # Filtrar "tanda" y valores no numÃ©ricos
        if 'tanda' in monto.lower() or monto.strip() == '':
            return None
        return float(monto.replace('$', '').replace(',', '').strip())
    return float(monto)

# Preparar datos de JosÃ© Luis
jl_montos = sorted([limpiar_monto(m) for m in jose_luis['Monto'].unique()])
mari_jl_montos = sorted([m for m in mari_jose_luis['monto_deuda'].unique()])

# Preparar datos de MarÃ­a Elena
me_valores = [limpiar_monto(m) for m in maria_elena['Monto'].dropna().unique()]
me_montos = sorted([m for m in me_valores if m is not None])
mari_me_montos = sorted([m for m in mari_maria_elena['monto_deuda'].unique()])

print("\n=== COMPARACIÃ“N - JOSÃ‰ LUIS ===")
print(f"\nMontos Ãºnicos registrados por JosÃ© Luis: {jl_montos}")
print(f"Montos Ãºnicos registrados por Mari: {mari_jl_montos}")

# Montos en Mari que NO estÃ¡n en JosÃ© Luis
mari_solo_jl = set(mari_jl_montos) - set(jl_montos)
print(f"\nâŒ Montos que Mari registra para JosÃ© Luis pero NO aparecen en su libreta: {sorted(mari_solo_jl)}")

# Montos en JosÃ© Luis que NO estÃ¡n en Mari
jl_solo = set(jl_montos) - set(mari_jl_montos)
print(f"âŒ Montos que JosÃ© Luis registra pero Mari NO tiene: {sorted(jl_solo)}")

print("\n" + "="*60)
print("=== COMPARACIÃ“N - MARÃA ELENA ===")
print(f"\nMontos Ãºnicos registrados por MarÃ­a Elena: {sorted(me_montos)}")
print(f"Montos Ãºnicos registrados por Mari: {mari_me_montos}")

# Montos en Mari que NO estÃ¡n en MarÃ­a Elena
mari_solo_me = set(mari_me_montos) - set(me_montos)
print(f"\nâŒ Montos que Mari registra para MarÃ­a Elena pero NO aparecen en su libreta: {sorted(mari_solo_me)}")

# Montos en MarÃ­a Elena que NO estÃ¡n en Mari
me_solo = set(me_montos) - set(mari_me_montos)
print(f"âŒ Montos que MarÃ­a Elena registra pero Mari NO tiene: {sorted(me_solo)}")
```

```python
print("\n\n" + "="*80)
print("DETALLE DE DISCREPANCIAS - JOSÃ‰ LUIS")
print("="*80)

print("\nğŸ“ MONTOS EN LA LIBRETA DE MARI QUE NO ESTÃN EN LA DE JOSÃ‰ LUIS:")
print("-" * 80)
for monto in sorted([2500, 4000, 7600, 12000]):
    registros = mari_jose_luis[mari_jose_luis['monto_deuda'] == monto]
    print(f"\n${monto:,}:")
    for idx, row in registros.iterrows():
        print(f"  â€¢ Fecha: {row['fecha_Aprox']}, Notas: {row['notas']}")

print("\n\n" + "="*80)
print("DETALLE DE DISCREPANCIAS - MARÃA ELENA")
print("="*80)

print("\nğŸ“ MONTOS EN LA LIBRETA DE MARI QUE NO ESTÃN EN LA DE MARÃA ELENA:")
print("-" * 80)
for monto in sorted([400, 500, 1000, 1200, 2500, 2640, 4000, 4400, 5600, 8600, 30000, 35000]):
    registros = mari_maria_elena[mari_maria_elena['monto_deuda'] == monto]
    print(f"\n${monto:,}:")
    for idx, row in registros.iterrows():
        print(f"  â€¢ Fecha: {row['fecha_Aprox']}, Notas: {row['notas']}")

print("\n\nğŸ“ MONTOS EN LA LIBRETA DE MARÃA ELENA QUE NO ESTÃN EN LA DE MARI:")
print("-" * 80)
for monto in [7600.0, 15000.0]:
    registros = maria_elena[maria_elena['Monto'].fillna('').apply(lambda x: str(x).replace('$', '').replace(',', '').strip() if isinstance(x, str) else str(x)) == str(int(monto))]
    print(f"\n${monto:,.2f}:")
    for idx, row in maria_elena.iterrows():
        try:
            monto_limpio = limpiar_monto(row['Monto'])
            if monto_limpio == monto:
                print(f"  â€¢ Fecha: {row['Fecha de prestamo']}, Abonos: {row['Abonos']}, Notas: {row['Notas']}")
        except:
            pass
```

```python
# CONSOLIDACIÃ“N: Combinar datos de JosÃ© Luis y MarÃ­a Elena
print("\n\n" + "="*100)
print("ANÃLISIS CONSOLIDADO - PRÃ‰STAMOS TOTALES DE TUS PADRES")
print("="*100)

# Consolidar registros de ambos padres
padres_consolidado = pd.concat([jose_luis, maria_elena], ignore_index=True)
print(f"\nTotal de registros de tus padres (JosÃ© Luis + MarÃ­a Elena): {len(padres_consolidado)}")

# Limpiar y preparar montos
padres_consolidado['Monto_limpio'] = padres_consolidado['Monto'].apply(
    lambda x: limpiar_monto(x) if isinstance(x, str) or isinstance(x, (int, float)) else None
)

# Eliminar valores None o NaN
padres_consolidado = padres_consolidado.dropna(subset=['Monto_limpio'])

# Consolidar registros de Mari
mari_consolidado = pd.concat([mari_jose_luis, mari_maria_elena], ignore_index=True)
print(f"Total de registros de Mari: {len(mari_consolidado)}")

# Agrupar por montos para cruce
print("\n" + "-"*100)
print("MONTOS TOTALES REGISTRADOS:")
print("-"*100)

padres_montos_count = padres_consolidado.groupby('Monto_limpio').size()
mari_montos_count = mari_consolidado.groupby('monto_deuda').size()

print("\nMONTOS DE TUS PADRES (cantidad de registros por monto):")
for monto, count in sorted(padres_montos_count.items()):
    print(f"  ${monto:>8,.2f}: {count} registro(s)")

print("\nMONTOS DE MARI (cantidad de registros por monto):")
for monto, count in sorted(mari_montos_count.items()):
    print(f"  ${monto:>8,.2f}: {count} registro(s)")
```

```python
# CRUCE DE PRÃ‰STAMOS: Identificar coincidencias y discrepancias
print("\n\n" + "="*100)
print("CRUCE DE PRÃ‰STAMOS: PADRES vs MARI")
print("="*100)

padres_montos_set = set(padres_consolidado['Monto_limpio'].unique())
mari_montos_set = set(mari_consolidado['monto_deuda'].unique())

# Montos que coinciden
coinciden = padres_montos_set & mari_montos_set
no_coinciden_padres = padres_montos_set - mari_montos_set
no_coinciden_mari = mari_montos_set - padres_montos_set

print(f"\nâœ… MONTOS QUE COINCIDEN (tus padres registran Y Mari tambiÃ©n): {len(coinciden)}")
for monto in sorted(coinciden):
    padres_count = (padres_consolidado['Monto_limpio'] == monto).sum()
    mari_count = (mari_consolidado['monto_deuda'] == monto).sum()
    print(f"   ${monto:>8,.2f}: Padres={padres_count} registro(s), Mari={mari_count} registro(s)")

print(f"\nâŒ MONTOS EN LIBRETA DE PADRES QUE NO ESTÃN EN LA DE MARI (posibles prestamos que Mari no registrÃ³): {len(no_coinciden_padres)}")
for monto in sorted(no_coinciden_padres):
    padres_count = (padres_consolidado['Monto_limpio'] == monto).sum()
    print(f"   ${monto:>8,.2f}: {padres_count} registro(s) - ESTOS NO DEBERÃA PAGAR (Mari no los registrÃ³)")

print(f"\nâš ï¸  MONTOS EN LIBRETA DE MARI QUE NO ESTÃN EN LA DE PADRES (posibles prestamos que padres no registraron): {len(no_coinciden_mari)}")
total_mari_no_identificado = mari_consolidado[mari_consolidado['monto_deuda'].isin(no_coinciden_mari)]['monto_deuda'].sum()
for monto in sorted(no_coinciden_mari):
    mari_count = (mari_consolidado['monto_deuda'] == monto).sum()
    print(f"   ${monto:>8,.2f}: {mari_count} registro(s)")
print(f"\n   TOTAL NO IDENTIFICADO EN PADRES: ${total_mari_no_identificado:,.2f}")
```

```python
# ANÃLISIS DETALLADO: Por quÃ© hay diferencias en montos
print("\n\n" + "="*100)
print("ANÃLISIS DE DISCREPANCIAS: Â¿Por quÃ© hay diferencias?")
print("="*100)

print("\n1ï¸âƒ£  PRÃ‰STAMO DE $15,000 EN PADRES (NO en Mari):")
print("-" * 100)
prestamo_15k = maria_elena[maria_elena['Monto'] == ' $15,000.00 ']
for idx, row in prestamo_15k.iterrows():
    print(f"   Fecha: {row['Fecha de prestamo']}")
    print(f"   Abonos: {row['Abonos']}")
    print(f"   Notas: {row['Notas']}")
    print(f"   CONCLUSIÃ“N: No estÃ¡ en libreta de Mari. Posible que sea un prÃ©stamo de otra persona.")

print("\n\n2ï¸âƒ£  MONTOS DE MARI NO REGISTRADOS EN PADRES - ANÃLISIS POR CATEGORÃA:")
print("-" * 100)

# Analizar montos pequeÃ±os (posibles intereses o comisiones)
montos_pequenos = [400, 500, 1000, 1200, 2500, 2640]
print(f"\nğŸ“Š MONTOS PEQUEÃ‘OS (posibles intereses/comisiones): ${sum([m for m in montos_pequenos if m in no_coinciden_mari]):,.2f}")
print(f"   Estos podrÃ­an ser: Intereses, comisiones, ajustes administrativos")
for m in montos_pequenos:
    if m in no_coinciden_mari:
        regs = mari_consolidado[mari_consolidado['monto_deuda'] == m]
        notas = regs['notas'].unique()
        print(f"   ${m:,}: {list(notas)}")

# Analizar montos grandes no identificados
montos_grandes = [4000, 4400, 5600, 8600, 12000, 30000, 35000]
total_grandes = sum([m for m in montos_grandes if m in no_coinciden_mari])
print(f"\nğŸ“Š MONTOS GRANDES NO IDENTIFICADOS: ${total_grandes:,.2f}")
print(f"   Estos podrÃ­an ser:")
print(f"   - PrÃ©stamos que los padres no registraron en sus libretas")
print(f"   - PrÃ©stamos de terceros (no a tus padres directamente)")
print(f"   - Errores de transcripciÃ³n o registros incompletos")
for m in montos_grandes:
    if m in no_coinciden_mari:
        regs = mari_consolidado[mari_consolidado['monto_deuda'] == m]
        notas = regs['notas'].unique()
        print(f"   ${m:,}: {list(notas)}")
```

```python
# RESUMEN EJECUTIVO Y CONCLUSIONES
print("\n\n" + "="*100)
print("âš ï¸  RESUMEN EJECUTIVO - CONCLUSIONES IMPORTANTES")
print("="*100)

# Calcular totales
padres_total = padres_consolidado['Monto_limpio'].sum()
mari_total = mari_consolidado['monto_deuda'].sum()
coinciden_total = sum([m * min((padres_consolidado['Monto_limpio'] == m).sum(), 
                                (mari_consolidado['monto_deuda'] == m).sum()) 
                       for m in coinciden])

print(f"\nğŸ’° TOTALES:")
print(f"   Total en libreta de PADRES: ${padres_total:,.2f}")
print(f"   Total en libreta de MARI: ${mari_total:,.2f}")
print(f"   Diferencia: ${abs(padres_total - mari_total):,.2f}")

print(f"\nâœ… PRÃ‰STAMOS CONFIRMADOS (aparecen en ambas libretas): ${coinciden_total:,.2f}")

print(f"\nâŒ PRÃ‰STAMOS SIN IDENTIFICAR EN PADRES (que Mari registra):")
no_id_total = mari_consolidado[mari_consolidado['monto_deuda'].isin(no_coinciden_mari)]['monto_deuda'].sum()
print(f"   Total: ${no_id_total:,.2f}")
print(f"   RecomendaciÃ³n: Verificar si estos son:")
print(f"      - Intereses o comisiones (montos pequeÃ±os)")
print(f"      - PrÃ©stamos a terceros (no a tus padres)")
print(f"      - Errores de registro")

print(f"\nâš ï¸  PRÃ‰STAMOS EN PADRES NO EN MARI:")
no_id_padres_total = padres_consolidado[padres_consolidado['Monto_limpio'].isin(no_coinciden_padres)]['Monto_limpio'].sum()
print(f"   Total: ${no_id_padres_total:,.2f}")
print(f"   Monto: $15,000 (MarÃ­a Elena) - Nota indica: 'Mayra tiene el control'")
print(f"   ğŸ¯ CONCLUSIÃ“N: NO DEBERÃAN PAGAR ESTO (Mari no lo registrÃ³)")

print("\n" + "="*100)
print("RECOMENDACIONES:")
print("="*100)
print("""
1. El prÃ©stamo de $15,000 NO debe ser pagado a Mari (no estÃ¡ en su libreta)
   
2. Los montos no identificados por tus padres ($119,740) deben verificarse:
   - Montos pequeÃ±os ($8,240): Probablemente intereses/comisiones
   - Montos grandes ($99,600): Posibles prÃ©stamos a terceros o errores
   
3. Los 9 montos que SÃ coinciden son los que definitivamente deben pagar

4. Discrepancia total entre libretas: ${abs(padres_total - mari_total):,.2f}
   Esta diferencia representa prÃ©stamos no documentados por ambas partes
""")
```

```python
print("\n\n" + "="*120)
print("TABLA DETALLADA: DESGLOSE DE MONTOS NO IDENTIFICADOS DE MARI")
print("="*120)

import pandas as pd

# Crear tabla de montos de Mari no identificados en padres
datos_no_id = []
for monto in sorted(no_coinciden_mari):
    regs_mari = mari_consolidado[mari_consolidado['monto_deuda'] == monto]
    for idx, row in regs_mari.iterrows():
        # Determinar a quiÃ©n se lo prestÃ³
        persona = row['etiqueta'].replace('libro_mari', '').strip()
        if 'jose' in str(row['notas']).lower():
            persona = "JosÃ© Luis"
        elif 'maria' in str(row['notas']).lower() or 'Elena' in str(row['notas']):
            persona = "MarÃ­a Elena"
        
        datos_no_id.append({
            'Monto': f"${monto:,.2f}",
            'Persona': persona,
            'Fecha': row['fecha_Aprox'],
            'Notas': row['notas'] if pd.notna(row['notas']) else 'sn'
        })

df_no_id = pd.DataFrame(datos_no_id)
print("\nğŸ“Œ MONTOS DE MARI NO REGISTRADOS EN LIBRETAS DE PADRES:")
print(df_no_id.to_string(index=False))

print("\n\n" + "-"*120)
print("DESGLOSE POR CATEGORÃA:")
print("-"*120)

print("\nğŸ”¹ POSIBLES INTERESES/COMISIONES (montos pequeÃ±os sin registros claros en padres):")
intereses_datos = []
for m in [400, 500, 1000, 1200, 2500, 2640]:
    if m in no_coinciden_mari:
        count = (mari_consolidado['monto_deuda'] == m).sum()
        intereses_datos.append({'Monto': f"${m:,}", 'Cantidad': count})

if intereses_datos:
    df_int = pd.DataFrame(intereses_datos)
    print(df_int.to_string(index=False))
    total_int = sum([m for m in [400, 500, 1000, 1200, 2500, 2640] if m in no_coinciden_mari])
    print(f"   SUBTOTAL: ${total_int:,.2f}")

print("\nğŸ”¹ MONTOS GRANDES SIN IDENTIFICAR (posibles prÃ©stamos no documentados):")
grandes_datos = []
for m in [4000, 4400, 5600, 8600, 12000, 30000, 35000]:
    if m in no_coinciden_mari:
        regs = mari_consolidado[mari_consolidado['monto_deuda'] == m]
        notas = ', '.join([str(n) for n in regs['notas'].unique() if pd.notna(n)])
        count = len(regs)
        grandes_datos.append({'Monto': f"${m:,}", 'Cantidad': count, 'Notas': notas if notas else 'Sin especificar'})

if grandes_datos:
    df_grd = pd.DataFrame(grandes_datos)
    print(df_grd.to_string(index=False))
    total_grd = sum([m for m in [4000, 4400, 5600, 8600, 12000, 30000, 35000] if m in no_coinciden_mari])
    print(f"   SUBTOTAL: ${total_grd:,.2f}")
```

```python
print("\n\n" + "="*120)
print("âœ… PRÃ‰STAMOS CONFIRMADOS (Aparecen en libretas de PADRES y MARI)")
print("="*120)

confirmados_data = []
for monto in sorted(coinciden):
    padres_count = (padres_consolidado['Monto_limpio'] == monto).sum()
    mari_count = (mari_consolidado['monto_deuda'] == monto).sum()
    
    # Obtener detalles del registro
    padres_reg = padres_consolidado[padres_consolidado['Monto_limpio'] == monto]
    mari_reg = mari_consolidado[mari_consolidado['monto_deuda'] == monto]
    
    # Sumar abonos
    padres_abonos = padres_reg['Abonos'].sum() if 'Abonos' in padres_reg.columns else 0
    mari_abonos = mari_reg['Abonos'].sum() if 'Abonos' in mari_reg.columns else 0
    
    # Calcular saldo
    saldo_padres = (monto * padres_count) - padres_abonos
    saldo_mari = (monto * mari_count) - mari_abonos
    
    confirmados_data.append({
        'Monto': f"${monto:,.2f}",
        'Registros (Padres/Mari)': f"{padres_count}/{mari_count}",
        'Total Prestado': f"${monto * padres_count:,.2f}",
        'Abonos (Padres)': f"${padres_abonos:,.2f}",
        'Saldo': f"${saldo_padres:,.2f}"
    })

df_confirmados = pd.DataFrame(confirmados_data)
print("\n" + df_confirmados.to_string(index=False))

total_prestado_confirmado = sum([m * min((padres_consolidado['Monto_limpio'] == m).sum(), 
                                          (mari_consolidado['monto_deuda'] == m).sum()) 
                                 for m in coinciden])
total_abonos_confirmado = sum([padres_consolidado[padres_consolidado['Monto_limpio'] == m]['Abonos'].sum() 
                               for m in coinciden])
total_saldo_confirmado = total_prestado_confirmado - total_abonos_confirmado

print("\n" + "="*120)
print(f"ğŸ’° TOTALES DE PRÃ‰STAMOS CONFIRMADOS:")
print(f"   Total Prestado (Confirmado): ${total_prestado_confirmado:,.2f}")
print(f"   Total Abonado: ${total_abonos_confirmado:,.2f}")
print(f"   SALDO PENDIENTE: ${total_saldo_confirmado:,.2f}")
print("="*120)
```

```python
print("\n\n" + "â–ˆ"*140)
print("ğŸ” ANÃLISIS DETALLADO: MONTOS DE MARI NO CRUZADOS CON PADRES")
print("â–ˆ"*140)

# Crear tabla de montos de Mari que NO estÃ¡n en padres
print("\n" + "="*140)
print("TABLA 1: MONTOS QUE MARI REGISTRA PERO NO APARECEN EN LIBRETA DE PADRES")
print("="*140)

datos_no_cruzados = []
id_monto = 1

# Obtener todos los montos Ãºnicos de Mari
montos_mari_unicos = sorted(set(mari_consolidado['monto_deuda'].unique()))

for monto in montos_mari_unicos:
    # Verificar si estÃ¡ en padres
    esta_en_padres = monto in padres_montos_set
    
    # Obtener registros de Mari para este monto
    regs_mari = mari_consolidado[mari_consolidado['monto_deuda'] == monto]
    
    for idx, row in regs_mari.iterrows():
        # Determinar a quiÃ©n se lo prestÃ³
        if 'jose' in str(row['notas']).lower():
            acreedor = "JosÃ© Luis"
        elif 'maria' in str(row['notas']).lower() or 'elena' in str(row['notas']).lower():
            acreedor = "MarÃ­a Elena"
        elif 'don jose' in str(row['notas']).lower():
            acreedor = "Don JosÃ© (Tercero?)"
        elif 'alma' in str(row['notas']).lower():
            acreedor = "Alma (Tercero?)"
        elif 'mayra' in str(row['notas']).lower():
            acreedor = "Mayra (Tercero?)"
        elif 'dora' in str(row['notas']).lower():
            acreedor = "Dora (Tercero?)"
        elif 'felix' in str(row['notas']).lower():
            acreedor = "Felix (Tercero?)"
        elif 'chave' in str(row['notas']).lower():
            acreedor = "Chave (Tercero?)"
        elif 'mireya' in str(row['notas']).lower():
            acreedor = "Mireya (Tercero?)"
        elif 'manuela' in str(row['notas']).lower():
            acreedor = "Manuela (Tercero?)"
        else:
            acreedor = "No especificado"
        
        # ClasificaciÃ³n
        if esta_en_padres:
            clasificacion = "âœ… IDENTIFICADO (En ambas libretas)"
        else:
            clasificacion = "âŒ NO CRUZADO (Solo en Mari)"
        
        datos_no_cruzados.append({
            'ID': id_monto,
            'Monto': f"${monto:,.0f}",
            'Acreedor/Deudor': acreedor,
            'Fecha': row['fecha_Aprox'] if pd.notna(row['fecha_Aprox']) else 'Sin fecha',
            'Notas': row['notas'] if pd.notna(row['notas']) else 'Sin notas',
            'Estado': clasificacion,
            'En Padres': 'SÃ' if esta_en_padres else 'NO'
        })
        id_monto += 1

df_no_cruzados = pd.DataFrame(datos_no_cruzados)

# Mostrar solo los NO CRUZADOS
df_no_cruzados_filtrado = df_no_cruzados[df_no_cruzados['En Padres'] == 'NO'].copy()

print("\n" + df_no_cruzados_filtrado.to_string(index=False))

# Resumen
print("\n" + "-"*140)
print("RESUMEN DE MONTOS NO CRUZADOS:")
print("-"*140)

total_no_cruzados = len(df_no_cruzados_filtrado)
total_monto_no_cruzados = df_no_cruzados_filtrado['Monto'].apply(lambda x: float(x.replace('$', '').replace(',', ''))).sum()

print(f"\nâœ‹ Total de registros NO CRUZADOS: {total_no_cruzados}")
print(f"ğŸ’° Monto total NO CRUZADO: ${total_monto_no_cruzados:,.0f}")

# Desglose por tipo
print("\nğŸ“Š DESGLOSE POR TIPO:")

# PrÃ©stamos a padres no registrados
posibles_padres = df_no_cruzados_filtrado[
    df_no_cruzados_filtrado['Acreedor/Deudor'].isin(['JosÃ© Luis', 'MarÃ­a Elena'])
]
if len(posibles_padres) > 0:
    monto_posibles_padres = posibles_padres['Monto'].apply(lambda x: float(x.replace('$', '').replace(',', ''))).sum()
    print(f"\n  1ï¸âƒ£  Posibles prÃ©stamos a PADRES (no registrados por ellos): {len(posibles_padres)} registros")
    print(f"     Monto total: ${monto_posibles_padres:,.0f}")
    for _, row in posibles_padres.iterrows():
        print(f"     â€¢ ID {row['ID']}: ${row['Monto']} - {row['Acreedor/Deudor']} - {row['Notas']}")

# Posibles prÃ©stamos a terceros
posibles_terceros = df_no_cruzados_filtrado[
    df_no_cruzados_filtrado['Acreedor/Deudor'].str.contains('Tercero', na=False) | 
    df_no_cruzados_filtrado['Acreedor/Deudor'].isin(['No especificado'])
]
if len(posibles_terceros) > 0:
    monto_posibles_terceros = posibles_terceros['Monto'].apply(lambda x: float(x.replace('$', '').replace(',', ''))).sum()
    print(f"\n  2ï¸âƒ£  Posibles prÃ©stamos a TERCEROS: {len(posibles_terceros)} registros")
    print(f"     Monto total: ${monto_posibles_terceros:,.0f}")
    for _, row in posibles_terceros.iterrows():
        print(f"     â€¢ ID {row['ID']}: ${row['Monto']} - {row['Acreedor/Deudor']} - {row['Notas']}")

# Intereses/comisiones (montos pequeÃ±os)
montos_pequeÃ±os_set = {400, 500, 1000, 1200, 1200, 2500, 2640}
posibles_intereses = df_no_cruzados_filtrado[
    df_no_cruzados_filtrado['Monto'].apply(lambda x: float(x.replace('$', '').replace(',', '')) in montos_pequeÃ±os_set)
]
if len(posibles_intereses) > 0:
    monto_posibles_intereses = posibles_intereses['Monto'].apply(lambda x: float(x.replace('$', '').replace(',', ''))).sum()
    print(f"\n  3ï¸âƒ£  Posibles INTERESES/COMISIONES: {len(posibles_intereses)} registros")
    print(f"     Monto total: ${monto_posibles_intereses:,.0f}")
    for _, row in posibles_intereses.iterrows():
        print(f"     â€¢ ID {row['ID']}: ${row['Monto']} - {row['Acreedor/Deudor']} - {row['Notas']}")

print("\n" + "â–ˆ"*140)
```

```python
# Exportar a CSV
csv_path = '/home/xinome/Documentos/Analisis_corregido/Montos_No_Cruzados_Mari.csv'
df_no_cruzados_filtrado.to_csv(csv_path, index=False, encoding='utf-8-sig')
print(f"\nâœ… Archivo CSV generado: {csv_path}")
print(f"   Total de filas: {len(df_no_cruzados_filtrado)}")

# Crear tabla adicional con informaciÃ³n de IDENTIFICACIÃ“N
print("\n\n" + "="*140)
print("TABLA 2: MATRIZ DE IDENTIFICACIÃ“N - TODOS LOS MONTOS (Cruzados y No Cruzados)")
print("="*140)

datos_matriz = []
id_matriz = 1

# Para cada monto Ãºnico de Mari
for monto in sorted(montos_mari_unicos):
    # Verificar si estÃ¡ en padres
    esta_en_padres = monto in padres_montos_set
    
    # Obtener registros de Mari
    regs_mari = mari_consolidado[mari_consolidado['monto_deuda'] == monto]
    count_mari = len(regs_mari)
    
    # Obtener registros de padres
    regs_padres = padres_consolidado[padres_consolidado['Monto_limpio'] == monto]
    count_padres = len(regs_padres)
    
    # Total monto
    total_monto = monto * max(count_mari, count_padres)
    
    # ClasificaciÃ³n
    if esta_en_padres:
        clasificacion = "âœ… IDENTIFICADO"
        estado = "Cruzado (en ambas)"
    else:
        clasificacion = "âŒ NO IDENTIFICADO"
        estado = "Solo en Mari"
    
    datos_matriz.append({
        'ID_Monto': id_matriz,
        'Monto': f"${monto:,.0f}",
        'Registros_Mari': count_mari,
        'Registros_Padres': count_padres,
        'Total_Monto': f"${total_monto:,.0f}",
        'Clasificacion': clasificacion,
        'Estado': estado
    })
    id_matriz += 1

df_matriz = pd.DataFrame(datos_matriz)
print("\n" + df_matriz.to_string(index=False))

# Exportar matriz
csv_matriz_path = '/home/xinome/Documentos/Analisis_corregido/Matriz_Identificacion_Montos.csv'
df_matriz.to_csv(csv_matriz_path, index=False, encoding='utf-8-sig')
print(f"\nâœ… Matriz de identificaciÃ³n exportada: {csv_matriz_path}")

# Resumen final
print("\n" + "-"*140)
print("RESUMEN FINAL DE IDENTIFICACIÃ“N:")
print("-"*140)

identificados = len(df_matriz[df_matriz['Clasificacion'].str.contains('IDENTIFICADO')])
no_identificados = len(df_matriz) - identificados

print(f"\nâœ… Montos IDENTIFICADOS (cruzados): {identificados}")
print(f"âŒ Montos NO IDENTIFICADOS (solo en Mari): {no_identificados}")
print(f"ğŸ“Š Total de montos Ãºnicos: {len(df_matriz)}")

total_identificado = df_matriz[df_matriz['Clasificacion'].str.contains('IDENTIFICADO')]['Total_Monto'].apply(
    lambda x: float(x.replace('$', '').replace(',', ''))
).sum()
total_no_identificado_matriz = df_matriz[~df_matriz['Clasificacion'].str.contains('IDENTIFICADO')]['Total_Monto'].apply(
    lambda x: float(x.replace('$', '').replace(',', ''))
).sum()

print(f"\nğŸ’° Monto IDENTIFICADO total: ${total_identificado:,.0f}")
print(f"ğŸ’° Monto NO IDENTIFICADO total: ${total_no_identificado_matriz:,.0f}")
```
